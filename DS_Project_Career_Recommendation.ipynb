### 1.1 Key Factors of the domain
When analyzing data science salaries in 2023, several key factors need to be considered to provide accurate and meaningful insights. These factors include:
*   **Work Year:** The number of hours or days an employee works per year, which can vary based on full-time, part-time, or contractual arrangements.
*   **Experience Level**: The number of years of relevant experience in the field of data science, which significantly influences salary levels.
*   **Employment Type:** Whether the position is full-time, part-time, contract-based, or freelance, as
different employment types offer varying salary structures and benefits.
*  **Employment Title:** The specific role within the data science domain, such as data scientist, data analyst,
machine learning engineer, or AI researcher, each with its own salary range.
*   **Salary USD:** The equivalent salary amount converted to US dollars, which provides a standardized comparison for salaries across different countries and currencies.
*   **Employment Location:** The location where the employment resides, as salaries can vary significantly based on different countries.

> Analyzing data science salaries in recent past years involves considering these factors in combination to provide a comprehensive understanding of the salary landscape in the field. Data scientists and analysts use statistical techniques and data modeling to extract meaningful insights from salary data, considering these variables and their interactions to draw accurate conclusions about salary
trends and patterns.

### 3.5 Employee Salary
The visualization of employees' salaries is necessary in order to identify any outliers within the dataset. To detect these outliers, it is essential to visualize salary column.




#### 3.5.a Detecting Outliers

##### Box Plot Before Removing Outliers

````python
sns.boxplot(x=df['salary_USD'], color = 'yellow')
````

> The box plot above illustrates that the majority of employee salaries fall under 200,000, while a few salaries are beyond the range of 300,000 that may become an outliers and if there are any outliers in our dataset, it is necessary to eliminate them by identifying the quantile in our dataset.













---
##### Removing Outliers

 **Finding  1st Quartile, 3rd Quartile and Interquartile Range**

````python
Q1 = df.quantile(0.25, numeric_only = True)
Q3 = df.quantile(0.75, numeric_only = True)
IQR = Q3 - Q1
print('Q1(quantile - 0.25) = ', Q1)
print('----------------------------------------------')
print('Q3(quantile - 0.75) = ', Q3)
print('----------------------------------------------')
print('IQR(Q3 - Q1) = ', IQR)
````



---


**Removing Outliers from dataframe**

````python
df_fix = df[~(((df < (Q1 - 1.5 *   IQR)) | (df > (Q3 + 1.5 *   IQR))).any(axis=1))]
df_fix.head(5)
````



---


**Dataset Shape before and after removing Outliers**

````python
#title Default title text
print("Dataset shape before removing Outliers :", df.shape)
print("Dataset shape before after Outliers :",  df_fix.shape)
print("Outliers Rows in dataset :", (df.size - df_fix.size))
df = df_fix
````

### Model Selection for salary prediction
We will be implementing several optimal machine learning models based on the specific needs of our project.

> To facilitate the advancement of our machine learning model for salary prediction in USD, we shall divide our dataframe into two distinct sets: train and test.

> I am going to set 75% of the data for the train set and 25% of the data for the test set by setting randome state to 42.

````python
scaler = StandardScaler()
df_scaler[['salary_USD']] = scaler.fit_transform(df_scaler[['salary_USD']])

X = df_scaler.drop(['salary_USD'], axis = 1)
y = df_scaler['salary_USD']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
X_train.shape, y_train.shape
````

#### Model - Linear Regression

````python
model = LinearRegression()

#fit to trainig data
model.fit(X_train, y_train)

# Predicting test data using model
y_pred = model.predict(X_test)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)

# Calculate the coefficient of determination (R^2 score)
r2 = r2_score(y_test, y_pred)

print(f'MSE = {mse},\nR^2 score = {r2}')
````

````python
dfr = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dfr.reset_index(drop=True, inplace=True)
dfr.head(10)
````

### Job Recommendation

````python
indices_exLevel = pd.Series(df['experience_level'])
indices_empTitle =  pd.Series(df['employment_title'])
def TitleMatching(title):
  if title not in indices_exLevel.values:
    if title not in indices_empTitle.values:
      return ''
    else:
      return 'employment_title'
  else:
    return 'experience_level'
````

````python
def Recommendation(searchBy):
  title = TitleMatching(searchBy)
  if title != '':
    tf = TfidfVectorizer(analyzer = "word", ngram_range=(1,2), min_df=0, stop_words='english')
    tfidf_matrix = tf.fit_transform(df[title])
    cosine_sim =  cosine_similarity(tfidf_matrix, tfidf_matrix)
    indices = pd.Series(df[title])
    idx = indices[indices == searchBy].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indices = list(score_series.iloc[1:11].index)
    return df.iloc[top_10_indices]
  else:
    return "Not found"
````

````python
result = Recommendation('ML Engineer')
if str(type(result)) != "<class 'str'>":
  idx = result.groupby(['employee_location','employment_title'])
  idxMax = idx['salary_USD'].idxmax()
  idxMin = idx['salary_USD'].idxmin()
  max_result = result.loc[idxMax]
  min_result = result.loc[idxMin]

  recommendationResult = '******** Job Recommendation ******** \n\n'
  for i in range(len(max_result)):
    obj_max = max_result.iloc[i]
    obj_min = min_result.iloc[i]
    salary_range = '$' + str(obj_max['salary_USD']) if (obj_max['salary_USD'] == obj_min['salary_USD']) else '$' + str(obj_min['salary_USD']) + ' - ' + '$' + str(obj_max['salary_USD'])
    recommendationResult += 'Job Title: ' + obj_max['employment_title'] + '\n' + 'Salary Range: '+ str(salary_range) + '\n' + 'Location: '+ obj_max['employment_location'] + '\n\n'
  print(recommendationResult)
````
